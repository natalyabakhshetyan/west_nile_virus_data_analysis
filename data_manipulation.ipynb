```python
import pandas as pd
```

## Unpivot years from wide to long format

```python
excel_file = 'West-Nile-Virus.xlsx'

# read the first sheet
total_cases_by_state = pd.read_excel(excel_file, sheet_name=1)
```

```python
total_cases_by_state.head()
```

```python
#drop Totals
total_cases_by_state = total_cases_by_state.drop('Total', axis = 1)
total_cases_by_state = total_cases_by_state[total_cases_by_state.State != 'Total']

total_cases_by_state.head()
```

```python
total_cases_by_state.tail()
```

```python
# melt columns to rows

total_cases_by_state = total_cases_by_state.melt(id_vars=["State"], 
                          var_name = "Year",
                         value_name = "Cases")

total_cases_by_state.head(50)
```

```python
# save to a new excel file

new_file = 'neuroinvasive_cases_by_state_melted.xlsx'
total_cases_by_state.to_excel(new_file, index = False, sheet_name = 'neuroinvasive_cases_by_state')
```
